__7. Faça um JOIN das duas tabelas, de forma que tenhamos apenas um dataset com as colunas "date", "higher_temp" e "lower_temp".__

```
#alterando o nome das colunas para melhorar a visualização do Join.
high_renamed = high_temps.withColumnRenamed('temp', 'temp_high')
low_renamed = low_temps.withColumnRenamed('temp', 'temp_low')
```

```
#criando tabelas temporárias.
high_renamed.createOrReplaceTempView('high')
low_renamed.createOrReplaceTempView('low')
```

Utilizado o sql para realizar o JOIN.

```
%sql 

SELECT
  h.*, l.temp_low
FROM 
  high as h
JOIN 
  low as l
ON 
  h.date == l.date
```

```
#criando um dataframe com o resultado do join.
new_data_temp = spark.sql('''
SELECT
  h.*, l.temp_low
FROM 
  high as h
JOIN 
  low as l
ON 
  h.date == l.date
''')
```

__8. Verifique o número de linhas da nova tabela e se há algum valor nulo (dica: o número de linhas deve ser idêntico ao número de linhas das tabelas separadas)__

__9. Crie uma coluna chamada "amplitude", que será higher_temp - lower_temp.__

```
new_data_temp = new_data_temp.withColumn('amplitude', F.col('temp_high') - F.col('temp_low'))

display(new_data_temp)
```

__10. Encontre os dias com a maior e a menor amplitude de temperatura.__

>> Menor amplitude.

```
from pyspark.sql.window import Window	
```

```
amp_min = new_data_temp.select('*').orderBy(F.col('amplitude').asc())
amp_min.select(amp_min.columns[0:4]).show(1)  
```

>>Maior amplitude.

```
amp_max = new_data_temp.select('*').orderBy(F.col('amplitude').desc())
amp_max.select(amp_max.columns[0:4]).show(1)  
```

__11. Crie uma coluna chamada "mean", que será a média da temperatura máxima e mínima de cada dia.__

```
columns = [F.col('temp_high'), F.col('temp_low')]
averageFunc = sum(x for x in columns) / len(columns)

data_with_mean = new_data_temp.withColumn('mean', averageFunc)

display(data_with_mean)
```

__12. Encontre os dias com a maior e a menor média de temperatura.__

>> Menor média de temperatura.

```
avg_temp_min = data_with_mean.select('*').orderBy(F.col('mean').asc())
avg_temp_min.select(avg_temp_min.columns[0:5]).show(1) 
```

>> Maior média de tmeperatura.

```
avg_temp_min = data_with_mean.select('*').orderBy(F.col('mean').desc())
avg_temp_min.select(avg_temp_min.columns[0:5]).show(1) 
```

__13. Encontre uma solução para particionarmos o dataset por ANO e MÊS. (dica: "yyyy-MM")__

__15. Agrupe o dataset por ANO e MÊS e encontre os seguintes valores agregados:__
- Temperatura máxima de cada mês
- Temperatura mínima de cada mês
- Temperatura média de cada mês
- Amplitude de cada mês
(dica: "yyyy-MM")

>> Leitura do dataset.

```
data_final = spark.read.parquet("/FileStore/tables/partition_date.parquet")

display(data_final)
```

>> Temperatura máxima de cada mês.

```
temp_max_per_month = data_final.groupby('date_partition')\
                               .max('temp_high')\
                               .withColumnRenamed('max(temp_high)', 'temp max per month')\
                               .orderBy(F.col('date_partition').asc())\
                               .show()
```

>> Temperatura mínima de cada mês.

```
temp_max_per_month = data_final.groupby('date_partition')\
                               .min('temp_low')\
                               .withColumnRenamed('min(temp_low)', 'temp min per month')\
                               .orderBy(F.col('date_partition').asc())\
                               .show()
```

>> Temperatura média de cada mês.

```
data_final.groupBy('date_partition')\
          .agg(F.round(F.mean('mean'), 2).alias('monthly average'))\
          .orderBy(F.col('date_partition').asc())\
          .show()
```

>> Amplitude de cada mês (dica: "yyyy-MM").

```
group = data_final.groupBy('date_partition')\
                  .sum('temp_high', 'temp_low')

group.withColumn('amplitude_mensal', F.col('sum(temp_high)') - F.col('sum(temp_low)'))\
     .drop('sum(temp_high)','sum(temp_low)')\
     .orderBy(F.col('date_partition').asc())\
     .show()
```
